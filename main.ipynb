```python
import cv2
from keras.models import load_model
from keras.preprocessing.image import load_img,img_to_array
import numpy as np
model=load_model(r'C:\\Users\hmisa\Desktop\Machine learning\Face Mask Detection Live Projects\model\model.h5')
img_width,imf_height=200,200
```

```python
import os
import cv2

cv2_base_dir = os.path.dirname(os.path.abspath(cv2.__file__))
haar_model = os.path.join(cv2_base_dir, 'data/haarcascade_frontalface_default.xml')
```

```python
face_cascade=cv2.CascadeClassifier(haar_model)
cap=cv2.VideoCapture(0)
img_count_full=0
font=cv2.FONT_HERSHEY_SIMPLEX
org=(1,1)
fontScale=1
color=(255,0,0)
thickness=2

while True:
    img_count_full+=1
    response,color_img=cap.read()
    
    if response==False:
        break
    
    
    # resize
    scale=50
    width=int(color_img.shape[1]*scale/100)
    height=int(color_img.shape[0]*scale/100)
    dim=(width,height)
    #resize image
    color_img=cv2.resize(color_img,dim,interpolation=cv2.INTER_AREA)
    
    # Convert to grayscale
    gray_img=cv2.cvtColor(color_img,cv2.COLOR_BGR2GRAY)
    
    # Detacts faces
    faces=face_cascade.detectMultiScale(gray_img,1.1,6)
    
    # take image and classify it
    img_count=0
    for(x,y,w,h) in faces:
        org=(x-10,y-10)
        img_count+=1
        color_face=color_img[y:y+h,x:x+w]
        cv2.imwrite(r'C:\\Users\hmisa\Desktop\Machine learning\Face Mask Detection Live Projects\faces\input/%d%dface.jpg'%(img_count_full,img_count),color_face)
        img=load_img(r'C:\\Users\hmisa\Desktop\Machine learning\Face Mask Detection Live Projects\faces\input/%d%dface.jpg'%(img_count_full,img_count),target_size=(img_width,imf_height))
        
        img=img_to_array(img)/255
        img=np.expand_dims(img,axis=0)
        pred_prob=model.predict(img)
        pred=np.argmax(pred_prob)
        
        if pred==0:
            print("User With Mask- Predic",pred_prob[0][0])
            class_label='mask'
            color=(255,0,0)
            cv2.imwrite(r'C:\\Users\hmisa\Desktop\Machine learning\Face Mask Detection Live Projects\faces\with_mask/%d%dface.jpg'%(img_count_full,img_count),color_face)
        else:
            print("User Without Mask- Predic",pred_prob[0][1])
            class_label='No_mask'
            color=(0,255,0)
            cv2.imwrite(r'C:\Users\hmisa\Desktop\Machine learning\Face Mask Detection Live Projects\faces\without_mask/%d%dface.jpg'%(img_count_full,img_count),color_face)
        cv2.rectangle(color_img,(x,y),(x+w,y+h),(0,0,255),3)
        
        cv2.putText(color_img,class_label,org,font,fontScale,color,thickness,cv2.LINE_AA)
    
    #display Image
    cv2.imshow('Live Face Mask Detection',color_img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
            
```
